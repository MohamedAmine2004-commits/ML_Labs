```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import os

# Fixe la graine pour les rsultats reproductibles
np.random.seed(42)

# Gnrer les annes d'exprience alatoires
years_of_experience = np.random.randint(1, 11, 40)

# Dfinir le salaire de base
base_salary = 2000

# Calculer le salaire en fonction des annes d'exprience
salary = base_salary + (years_of_experience * 200) + np.random.randint(1, 401, 40)

# Crer un DataFrame
data = pd.DataFrame({
    'Years': years_of_experience,
    'Salary': salary
})

# Spcifier le rpertoire du bureau
directory = 'C:\\Users\\el badr info\\Desktop'  # Remplacez YourUsername par votre nom d'utilisateur

# Crez le rpertoire s'il n'existe pas
if not os.path.exists(directory):
    os.makedirs(directory)

csv_path = os.path.join(directory, 'employee_salaries.csv')

# Enregistrer le DataFrame dans un fichier CSV
data.to_csv(csv_path, index=False)

# Charger les donnes  partir du fichier CSV
loaded_data = pd.read_csv(csv_path)

# Tracer les donnes
plt.scatter(loaded_data['Years'], loaded_data['Salary'], color='blue', label='Data points')
plt.title('Salary vs. Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

# Prparer les donnes pour la rgression linaire
X = loaded_data['Years'].values.reshape(-1, 1)
y = loaded_data['Salary']

# Crer et entraner le modle de rgression linaire
model = LinearRegression()
model.fit(X, y)

# Prdire le salaire pour 10 annes d'exprience
predicted_salary = model.predict([[10]])
print(predicted_salary)

# Tracer la ligne de rgression
plt.scatter(loaded_data['Years'], loaded_data['Salary'], color='blue', label='Data points')
plt.plot(loaded_data['Years'], model.predict(X), color='red', label='Regression line')
plt.title('Salary vs. Years of Experience with Linear Regression')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()


```
