```python
def fix_memory(item):
    if 'TB' in item:
        if '2' in item:
            return 2000
        else:
            return 1000
    else:
        gb = int(item.split('G')[0])
        return gb
```

```python


def total_memory(item):
    # step 1 is to check if there is a + sign
    if '+' in item:
        # if ture then we have to do things twice
        sum = 0 
        x = item.split('+')
        sum += fix_memory(x[0])
        sum += fix_memory(x[1])
    else:
        x = item.split()[0]
        sum =fix_memory(x)
    return sum 
        

```

```python
cat_x = ['Company', 'Product', 'TypeName','RAM (GB)', 'GPU_Company', 'GPU_Type', 'OpSys','BaseScreenResolution']
x_train[cat_x]
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler


# Custom wrapper for LabelEncoder
class LabelEncoderTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.encoder = None

    def fit(self, X, y=None):
        self.encoder = {col: LabelEncoder().fit(X[col]) for col in X.columns}
        return self

    def transform(self, X):
        X_transformed = X.copy()
        for col in X.columns:
            X_transformed[col] = self.encoder[col].transform(X[col])
        return X_transformed
```

```python
num_list =list(x_num)
full_pipeline = ColumnTransformer([
    ('label',LabelEncoderTransformer(),cat_x),
    ('num',StandardScaler(),num_list)
])
x_prep = full_pipeline.fit_transform(x_train)
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x_prep, y_train)
```

```python
from sklearn.metrics import mean_squared_error
reg_predictions = lin_reg.predict(x_prep)
reg_predictions
```

```python
np.sqrt(mean_squared_error(reg_predictions,y_train))
```
